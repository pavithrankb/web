```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from datetime import datetime
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.layers.convolutional import Conv1D, MaxPooling1D
```

```python
df = pd.read_csv("cleaned_finalset.csv", header=0, index_col=0)
df.describe()
```

```python
dataset_columns = df.columns.tolist()
dataset_columns
```

```python
df = df.drop(['From Date','To Date',
 'AT',
 'BP',
 'PM2.5',
 'RH',
 'WD',
 'WS',
 'CO',
 'NH3',
 'NO',
 'NO2',
 'NOx',
 'SO2',
 'Ozone'
 ],axis = 1)
df.head()
```

```python
df
```

```python
def split_seq(sequence, steps):
  x,y = list(), list()
  for start in range(len(sequence)):
    end=start+steps
    if end>len(sequence)-1:
      break
    seqx,seqy = sequence[start : end], sequence[end]
    x.append(seqx)
    y.append(seqy)
  return np.array(x),np.array(y)
```

```python
actual_seq=list(df['PM10'])
```

```python
sample_seq=actual_seq
```

```python
steps= 1000
x,y = split_seq(sample_seq , steps)
print("----------------")
print("raw")
print(sample_seq)
print("features")
print(x)
print("output")
print(y)
print("--------------")
later_use_x=x
later_use_y=y
```

```python
features=1
x=x.reshape((x.shape[0] , x.shape[1] , features))
print("after reshaping")
print(x.shape)
print("------------")
print(x)
```

```python
model = Sequential()
model.add(Conv1D(filters =64 , kernel_size=2 , activation ='relu', input_shape=(steps,features)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(1))
model.compile(optimizer= 'adam', loss='mse')
```

```python
model.fit(x,y,epochs=100,verbose=0)
```

```python
x_input=later_use_x
x_input = x_input.reshape((x.shape[0] , x.shape[1] , features))
y_pred= model.predict(x_input, verbose=0)
print(len(y_pred))
print(y_pred)
```

```python
print(len(later_use_y))
print(later_use_y[200:210])
plt.plot(later_use_y)
plt.plot(y_pred, color='red')
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(later_use_y,y_pred)
```
